```python
import spotify

sc = spotify.SpotifyClient()

token = sc.authenticate()
# print("Access token:", token)
```

```python
from spotify import SearchResult, SearchType

res: SearchResult = sc.search(
    "Bohemian Rhapsody",
    SearchType.TRACK,
)

# Get the first track from the search results
tracks = res["tracks"]["items"]
for track in tracks:
    print(
        f"Track name: {track['name']}, Artist: {track['artists'][0]['name']}, URI: {track['uri']}"
    )
```

```python
# Creating a playlist requires a user OAuth token (not client credentials).
# Make sure http://127.0.0.1:8080/callback is added as a Redirect URI
# in your Spotify app dashboard at https://developer.spotify.com/dashboard

user_token = sc.authorize_user()  # opens browser, waits for approval

playlist = sc.create_playlist(
    name="My New Playlist",
    description="A playlist created via the Spotify API",
    public=False,
    user_access_token=user_token,
)
print(f"Created: {playlist['name']} — {playlist['external_urls']['spotify']}")
```
